```python
#1
import pandas as pd
dataframe = pd.read_csv("/content/drive/MyDrive/AI_SEM3/Datasets/Titanic-Dataset.csv")

# Add a new column 'fareperyear' which is Fare divided by Age
dataframe['fare_per_year'] = dataframe['Fare'] / dataframe['Age']
# Add a new column 'fareperclass' (Fare divided by Pclass)
dataframe['fareperclass'] = dataframe['Fare'] / dataframe['Pclass']
dataframe.head()
```

```python
#2 Subset rows for adult male passengers (Age >= 18 and Sex == "male")
adultmales = dataframe[(dataframe['Sex'] == 'male') & (dataframe['Age'] >= 18)]
adultmales.head()
```

```python
#3 Sort adultmales by 'fareperclass' in descending order
adult_males_srt = adultmales.sort_values(by='fareperclass', ascending=False)
adult_males_srt.head()
```

```python
#4 Select only 'Name', 'Age', and 'fareperclass' columns
result = adult_males_srt[['Name', 'Age', 'fareperclass']]
result.head()
```

 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
 Based on the dataset Answer the following question:

 1. Calculate the total Fare paid across all passengers in the Titanic dataset.
 2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
 3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).
 4. Combine the fare totals from first, second, and third classes into a list.
 5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.


```python
#1 Total Fare paid across all passengers
total_fare = dataframe['Fare'].sum()
print(f"Total Fare paid across all passengers in the Titanic dataset: {total_fare}")
```

```python
#2 In first class (Pclass is 1)
first_class_fare = dataframe[dataframe['Pclass'] == 1]['Fare'].sum()
print(f"Total Fare paid by passengers in first class: {first_class_fare}")
```

```python
#3 In  second class (Pclass is 2) and third class (Pclass is 3)
second_class_fare = dataframe[dataframe['Pclass'] == 2]['Fare'].sum()
print(f"Total Fare paid by passengers in second class: {second_class_fare}")

third_class_fare = dataframe[dataframe['Pclass'] == 3]['Fare'].sum()
print(f"Total Fare paid by passengers in third class: {third_class_fare}")
```

```python
#4 Combine the fare totals from first, second, and third classes into a list.
fare_totals = [first_class_fare, second_class_fare, third_class_fare]
print(f"Fare totals for each class: {fare_totals}")
```

```python
#5 The proportion of fare revenue by class
fare_proportions = [first_class_fare / total_fare, second_class_fare / total_fare, third_class_fare / total_fare]
print(f"Proportion of fare revenue by class: {fare_proportions}")
```
