Task 4. Analyze Temperature Patterns by Time of Day:

2. Iterate over the temperatures list, assigning values to each time-of-day list based on

4. (Optional) Plot ”day vs. temperature” using matplotlib.

```python
import matplotlib.pyplot as plt

# Create empty lists for night, day, and evening temperatures
night = []       # Temperatures from 00:00 to 08:00
evening = []     # Temperatures from 08:00 to 16:00
day = []         # Temperatures from 16:00 to 24:00

temperatures = [23, 24, 25, 26, 27, 28, 29, 30, 30, 28, 5, 7, 9, 11, 13, 26, 23, 21, 19, 15, 17, 19, 21, 22]
#Iterate over the temperatures list, assigning values to each time-of-day list based on their position.
for hour, temp in enumerate(temperatures):
    if 0 <= hour < 8:
        night.append(temp)
    elif 8 <= hour < 16:
        evening.append(temp)
    elif 16 <= hour < 24:
        day.append(temp)

#Calculate and print the average day-time temperature
average_day_temperature = sum(day) / len(day) if day else 0
print(f"Average daytime temperature: {average_day_temperature:.2f}°C")

# Plot "day vs. temperature" using matplotlib
hours = list(range(16, 24))  # Hours from 16:00 to 24:00
plt.plot(hours, day, marker='o', color='b', label="Day Temperature")
plt.xlabel("Hours")
plt.ylabel("Temperature (°C)")
plt.title("Day vs. Temperature")
plt.grid(True)
plt.legend()
plt.show()
```

Task 1 - Sum of Nested Lists:
Scenario: You have a list that contains numbers and other lists of numbers (nested lists).
You want to find the total sum of all the numbers in this structure.
Task:
• Write a recursive function sum_nested_list(nested_list) that:
1. Takes a nested list (a list that can contain numbers or other lists of numbers) as
input.
2. Sums all numbers at every depth level of the list, regardless of how deeply nested
the numbers are.
• Test the function with a sample nested list, such as
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]].
The result should be the total sum of all the numbers.


```python
def sum_nested_list(nested_list):
    total = 0
    for item in nested_list:
        if isinstance(item, list):
            total += sum_nested_list(item)
        else:
            total += item
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)

print("Total sum of the nested list:", result)
```

Task 2 - Generate All Permutations of a String:
Scenario: Given a string, generate all possible permutations of its characters. This is useful
for understanding backtracking and recursive depth-first search.
Task:
• Write a recursive function generate_permutations(s) that:
– Takes a string s as input and returns a list of all unique permutations.
• Test with strings like ”abc” and ”aab”.
print(generate_permutations("abc"))


```python
def permutations(s):
    def backtracking(start, end):
        if start == end:
            permutations.append(''.join(s))
        for i in range(start, end):
            if i > start and s[i] == s[start]:
                continue
            s[start], s[i] = s[i], s[start]  # Swap characters at start and i
            backtracking(start + 1, end)  # Recurse for the next position
            s[start], s[i] = s[i], s[start]  # Backtrack and restore the original order

    s = list(s)
    permutations = []
    s.sort()  # Sort the string to handle duplicates efficiently
    backtracking(0, len(s))
    return permutations

# Test with string
print("Permutations of abc:", permutations("abc"))
print("Permutations of aab:", permutations("aab"))
```

Task 3 - Directory Size Calculation:
Directory Size Calculation Scenario: Imagine a file system where directories can contain files
(with sizes in KB) and other directories. You want to calculate the total size of a directory,
including all nested files and subdirectories.

1. Write a recursive function calculate_directory_size(directory) where:
• directory is a dictionary where keys represent file names (with values as sizes in
KB) or directory names (with values as another dictionary representing a subdirectory).
• The function should return the total size of the directory, including all nested
subdirectories.
2. Test the function with a sample directory structure.

```python
def calculate_directory_size(directory):
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value

    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test the function
print("The total size of the directory is",calculate_directory_size(directory_structure))
```

8.2.2 Exercises - Dynamic Programming:
Task 1 - Coin Change Problem:
Scenario: Given a set of coin denominations and a target amount, find the minimum number
of coins needed to make the amount. If it’s not possible, return - 1.
Task:
1. Write a function min_coins(coins, amount) that:
• Uses DP to calculate the minimum number of coins needed to make up the
amount.
2. Test with coins = [1, 2, 5] and amount = 11. The result should be 3 (using coins
[5, 5, 1]).


```python
def min_coins(coins, amount):

    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

#Task2
coins = [1,2,5]
amount=11
result = min_coins(coins, amount)
print(f"The minimum number of coins{coins} needed to make the amount {amount} is",result)
```

Task 2 - Longest Common Subsequence (LCS):
Scenario: Given two strings, find the length of their longest common subsequence (LCS).
This is useful in text comparison.
Task:
1. Write a function longest_common_subsequence(s1, s2) that:
• Uses DP to find the length of the LCS of two strings s1 and s2.
2. Test with strings like "abcde" and "ace"; the LCS length should be 3 ("ace").

```python
def longest_common_subsequence(s1, s2):
    # Initialize a DP table with dimensions (len(s1)+1) x (len(s2)+1)
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]

    # Fill the DP table
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1  # Characters match
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # The bottom-right corner of the table contains the length of the LCS
    return dp[len(s1)][len(s2)]

# Test with strings
s1 = "abcde"
s2 = "ace"
print(f"Length of LCS between '{s1}' and '{s2}':", longest_common_subsequence(s1, s2))
```

Task 3 - 0/1 Knapsack Problem:
Scenario: You have a list of items, each with a weight and a value. Given a weight capacity,
maximize the total value of items you can carry without exceeding the weight capacity.
Task:
1. Write a function knapsack(weights, values, capacity) that:
• Uses DP to determine the maximum value that can be achieved within the given
weight capacity.

2. Test with weights [1, 3, 4, 5], values [1, 4, 5, 7], and capacity 7. The re-
sult should be 9.

```python
def knapsack(weights, values, capacity):
    n = len(weights)  # Number of items
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):  # Loop through items
        for w in range(1, capacity + 1):  # Loop through each weight capacity
            if weights[i - 1] <= w:
                # Item can be included: take max of including or excluding the item
                dp[i][w] = max(dp[i - 1][w], values[i - 1] + dp[i - 1][w - weights[i - 1]])
            else:
                # Item cannot be included: value is same as without this item
                dp[i][w] = dp[i - 1][w]

    # The bottom-right corner of the table contains the maximum value
    return dp[n][capacity]

# Test the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(f"Maximum value for capacity {capacity}:", knapsack(weights, values, capacity))
```
