 Task- 3- Outlier Detection:

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
 2. Define outliers using the 1.5 × IQR rule.

```python
def detect_countries_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers  # Ensure this is the last line in the function

# Define the columns to check
columns_to_check = [
    'score',
    'Log GDP per capita',
    'composite score',
]

# Loop through each column to detect and print outliers
for column in columns_to_check:
    print(f"\nOutliers for column: {column}")
    outliers = detect_countries_iqr(southAsian_df, column)
    if not outliers.empty:
        print(outliers[['Country name', column]])
    else:
        print("No outliers found.")
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
 in a different color.

```python
import matplotlib.pyplot as plt
outliers_gdp = detect_countries_iqr(southAsian_df, 'Log GDP per capita')
outliers_score = detect_countries_iqr(southAsian_df, 'score')
plt.figure(figsize=(10, 5))

plt.scatter(southAsian_df['Log GDP per capita'], southAsian_df['score'], label='Countries', color='green')
plt.scatter(outliers_gdp['Log GDP per capita'], outliers_gdp['score'], label='GDP Outliers', color='blue', marker='x')
plt.scatter(outliers_score['Log GDP per capita'], outliers_score['score'], label='Score Outliers', color='red', marker='x')

plt.xlabel('Log GDP per Capita', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.title('GDP per Capita vs. Score with Outliers Highlighted', fontsize=14)

plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
